```python
#실습 1. 아래와 같은 그래프를 그려보세요.
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_2019 = [100,120,140,110,130,150,160,170,180,200,190,210]
sales_2020 = [90,110,130,120,140,160,170,160,150,180,200,190]

plt.plot(months, sales_2019, label='2019')
plt.plot(months, sales_2020, label='2020')
plt.title('Monthly Sales Comparison (2019-2020)')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
#실습2. 아래와 같은 그래프를 그려보세요.
import matplotlib.pyplot as plt
import numpy as np

categories = ['Category 1','Category 2','Category 3','Category 4','Category 5']
data = [20, 35, 15, 27, 45]

x = np.arange(5)
plt.bar(categories, data)
plt.xticks(rotation=45)
plt.yticks([0,10,20,30,40,50] )
plt.title("Bar Chart")
plt.grid(True, axis='both') 
plt.xlabel('categories')
plt.ylabel('Values')
plt.show()
```

```python
#실습3. 아래와 같은 그래플 그려보세요.
labels = ['Apple','Banana','Mango','Blueberry']
size = [15,30,45,10]
explode = (0, 0.1, 0, 0) #Banana 부분에서만 강조

plt.pie(size, labels=labels, autopct='%.1f%%', explode=explode, startangle=90, shadow=90)
plt.show()
```

```python
#실습4. sin&cos 함수 그리기
x = np.linspace(0, 10, 100)

# sin과 cos 함수의 값을 계산
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x,y_sin,color='blue',label='sin(x)')
plt.plot(x,y_cos,color='red',label='cos(x)')
plt.title('Sine and Cosine functions')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True, axis='both') 

plt.legend()
plt.show()
```

```python
import folium
import pandas as pd

# CSV 파일 경로 설정
file_path = r"C:\Users\zer12\Desktop\git-test\택시승차대 현황.csv"

# 1. CSV 파일 읽기 (인코딩 지정)
data = pd.read_csv(file_path, encoding='CP949')  # 또는 encoding='EUC-KR'

# 2. 지도 생성 (마포구 중심 좌표 설정)
m = folium.Map(location=[37.541, 126.948], zoom_start=14)

# 3. CSV 데이터 기반으로 마커 추가
for index, row in data.iterrows():
    folium.Marker(
        [row['위도'], row['경도']],  # '위도'와 '경도' 열을 사용
        popup=row['위치명'],  # '위치명' 열을 팝업에 표시
        tooltip=f"{row['위치명']} 위치"  # 툴팁으로도 '위치명' 표시
    ).add_to(m)

# 4. 지도 저장
m.save("map_with_taxi_stands.html")

# 5. Jupyter Notebook에서 실행 시 지도 표시
m
```
