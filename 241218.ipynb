```python
font1 = {'family' : 'fantasy', 'color' : 'deeppink', 'weight' : 'normal', 'size':'xx-large'}
plt.plot([1,2,3,4], [1,4,9,16])
plt.xlabel('x', labelpad=30, loc='left', fontdict=font1)
plt.ylabel('y',loc='top')
plt.show()
```

```python
line1, = plt.plot([1,2,3], label='label1')
line2, = plt.plot([2,4,6], label='label2')
line3, = plt.plot([3,6,9], label='label3')
# plt.legend(handles=[line1,line2,line3], loc=(1,0))
# plt.legend()
plt.legend(ncol=3, fontsize=14, frameon=True, shadow=True)
plt.title("legend Test")
plt.grid(True, axis='x') #axis=both
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)
y = [100,400,900]

plt.bar(x, [100,400,900], color=['r','g','b'], width=0.6)
plt.plot(x,y)
plt.xticks(x, ['2018', '2019', '2020'])
# plt.yticks(y, ['a', 'b', 'c'])
plt.show()
```
