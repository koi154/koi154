```python
import pandas as pd


company_names = ['S-Oil', 'SK텔레콤', '삼성화재', '넥슨코리아', '셀트리온', '한미사이언스', '홈플러스', '롯데쇼핑', '메가스터디']
file_path = r'C:\Users\zer12\Desktop\git-test\연봉별 퇴사율 조사.xlsx'
sheets = pd.read_excel(file_path, sheet_name=company_names)

for company, df in sheets.items():
    print(f"Sheet: {company}")
    print(df)  # 각 시트의 상위 데이터 출력
```

```python
# Step 1: 각 시트에서 필요한 열이 있는지 확인
for company, df in sheets.items():
    print(f"{company}")
    print(f"{df.columns}\n")
```

```python
# Step 2: 필요한 열만 추출
cleaned_data = []
for company, df in sheets.items():
    df = df[['년도','종업원수', '퇴사자수']].dropna() # 결측값 제거
    print(F"{company}")
    print(df, "\n")
    cleaned_data.append((company, df))
```

```python
# Step 3: 퇴사율 계산
processed_data = []
for company, df in sheets.items():
    df = df[['년도', '종업원수', '퇴사자수']].dropna()
    df['퇴사율'] = (df['퇴사자수'] / df['종업원수']) * 100 # 퇴사율 계산
    df['회사명'] = company # 회사명 추가
    print(f"{company}")
    print(df, "\n")
    processed_data.append(df)
```
