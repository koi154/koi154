```python
from bs4 import BeautifulSoup

html_str = ''' # < 여러줄을 str 로 만들 떄 사용
<html>
    <body>
        <div id="content">
            <ul class = 'industry'>
                <li>인공지능</li>
                <li>빅데이터</li>
                <li>스마트팩토리</li>
            </ul>
            <ul class = 'comlang'>
                <li>Python</li>
                <li>C++</li>
                <li>Javascript</li>
            </ul>
        </div>
    </body>
</html>
'''
soup = BeautifulSoup(html_str, "html.parser") # html 파싱
```

```python
# 처음 나오는 ul 태그로 찾기
first_ul = soup.find('ul')
print(first_ul)
print(first_ul.text) # 태그 없이 텍스트 출력
print(first_ul.get_text)
```

```python
all_list = first_ul.findAll('li')
print(all_list) # [<li>인공지능</li>, <li>빅데이터</li>, <li>스마트팩토리</li>]
print(all_list[0]) # <li>인공지능</li>
print(all_list[0].text) #인공지능

for li in all_list:
    print(li.text) # 인공지능 빅데이터 스마트팩토리
```

```python
# class 선택자로 찾기
second_ul = soup.find('ul', class_='comlang')
print(second_ul) # <ul class="comlang">, <li>Python</li>, <li>C++</li>, <li>Javascript</li>, </ul>
print(second_ul.text) #Python, C++, Javascript
```

```python
#select_one 사용
f_ul = soup.select_one('ul.industry')
print(f_ul)
print(f_ul.text)
```

```python
#실습
import requests
from bs4 import BeautifulSoup

query = input("검색어를 입력하세요: ") # 클래스 쿼리 입력하기


response = requests.get(f"https://search.naver.com/search.naver?where=news&ie=utf8&sm=nws_hty&query={query}") # 뉴스 링크 가져오기
html = response.text # 뉴스 텍스트화 하기

soup = BeautifulSoup(html, 'html.parser') # Html 파싱

news_items = soup.select(".news_tit") # 뉴스 제목 클래스를 리스트화

if news_items: #뉴스 있는 경우
    for item in news_items: 
        title = item.text # 아이템을 텍스트화
        link = item['href'] # 링크 추출
        print("제목:", title)
        print("링크:", link)
        # print("-"*50) # 구분선
else: #뉴스가 없는 경우
    print("뉴스를 찾을 수 없습니다.")
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://finance.naver.com/marketindex/"

response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser') # Html 파싱

exchange_rates = soup.select("ul.data_lst li")[:4]

for rate in exchange_rates:
    currency = rate.select_one("h3 > span").text.strip()
    value = rate.select_one(".value").text.strip()
    print(f"{currency.split()[1]}: {value}")
```

```python
import requests
from bs4 import BeautifulSoup
import openpyxl

# 1. 엑셀 파일 생성
wb = openpyxl.Workbook()
ws = wb.active
ws.title = "환율 정보"  # 워크시트 이름 설정

# 2. 헤더 추가
ws.append(["통화", "환율"])  # 첫 번째 줄에 헤더 추가

# 3. 크롤링 데이터 가져오기
url = "https://finance.naver.com/marketindex/"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# 환율 정보 선택 (상위 4개 항목만 가져오기)
exchange_rates = soup.select("ul.data_lst li")[:4]

for rate in exchange_rates:
    currency = rate.select_one("h3 > span").text.strip().split()[-1]  # 통화명
    value = rate.select_one(".value").text.strip()  # 환율 값
    ws.append([currency, value])  # 데이터를 엑셀에 추가

# 4. 엑셀 파일 저장
wb.save("exchange_rates.xlsx")
print("엑셀 파일 저장 완료!")
```
